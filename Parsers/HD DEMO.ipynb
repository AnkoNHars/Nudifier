```python
import skimage.io
import numpy as np
from PIL import Image
```

```python
def simple_panorama(i):
    full = skimage.io.imread('data/256x256/test/{}.png'.format(i))
    pose = skimage.io.imread('data/256x256/cropped/test/face/{}.png'.format(i))
    nude = skimage.io.imread('../../Pix2pixHD/pix2pixHD-master/results/dressed2nude/test_latest/images/{}_synthesized_image.jpg'.format(i))
    nude[pose == np.repeat(0,3)] = full[pose == np.repeat(0,3)]
    panorama = np.concatenate([full, nude], axis = 1)
    panorama = Image.fromarray(panorama, 'RGB')
    return panorama
```

```python
simple_panorama(28)
```
